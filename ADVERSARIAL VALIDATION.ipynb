# Adversarial Validation — Detecting Train/Test Distribution Shift

<div style='background:linear-gradient(135deg,#0f172a 0%,#1a2744 50%,#1d4ed8 100%);border-radius:12px;padding:24px 28px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#fff;line-height:1.7;margin-bottom:8px;'>
<b style='font-size:18px;letter-spacing:0.4px;'>Notebook overview</b><br><br>
<b>Adversarial validation</b> is a technique for detecting whether your training and test sets come from the same distribution. If they don't, models trained on the training set may generalise poorly — adversarial validation helps you <em>quantify</em> and <em>diagnose</em> that problem before it costs you on the leaderboard.<br><br>
<b style='color:#93c5fd;'>Core idea:</b> Combine train and test, label them (train=0, test=1), train a classifier to tell them apart. If it succeeds (AUC ≫ 0.5), your sets look different. If it fails (AUC ≈ 0.5), they're indistinguishable — good news.
<br><br>

## Section 2 — Creating Synthetic Data with Distribution Shift

<div style='background:#fffbeb;border-left:6px solid #d97706;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;'>
<b style='color:#92400e;font-size:15px;'>Simulating a realistic temporal shift scenario</b><br><br>
We generate a controlled dataset where <b>3 out of 10 features</b> have drifted between the train and test collection periods. This mimics common real-world problems:
<ul style='margin-top:8px;padding-left:18px;line-height:2;'>
  <li><b>Seasonal shifts</b> — user behaviour or environmental signals that change over months</li>
  <li><b>Data pipeline bugs</b> — a preprocessing step applied inconsistently to train vs test</li>
  <li><b>Population drift</b> — the underlying population the model is deployed on has changed</li>
</ul>
The 7 unshifted features are drawn from the same distribution in both sets — making the adversarial classifier's job harder and the identified shifted features more reliable signal.
</div>

```python
def make_shifted_data(n_train=2000, n_test=800, n_features=10, n_shifted=3):
    """
    Generate train/test data where `n_shifted` features have distribution shift.
    
    Returns:
        X_train, X_test : feature DataFrames
        y_train         : target labels (for the original ML task)
    """
    feature_names = [f"feature_{i}" for i in range(n_features)]
    
    # --- Train set ---
    X_train = pd.DataFrame(
        np.random.randn(n_train, n_features),
        columns=feature_names
    )
    
    # --- Test set: most features same, but `n_shifted` are different ---
    X_test = pd.DataFrame(
        np.random.randn(n_test, n_features),
        columns=feature_names
    )
    
    shifted_features = feature_names[:n_shifted]
    for feat in shifted_features:
        X_test[feat] = X_test[feat] * 2.5 + 3.0   # scale & shift
    
    # Target variable (only exists for train)
    y_train = (X_train["feature_3"] + np.random.randn(n_train) > 0).astype(int)
    
    return X_train, X_test, y_train, shifted_features

X_train, X_test, y_train, shifted_features = make_shifted_data(
    n_train=2000, n_test=800, n_features=10, n_shifted=3
)

print(f"Train shape : {X_train.shape}")
print(f"Test shape  : {X_test.shape}")
print(f"Shifted features (ground truth): {shifted_features}")
```

## Section 3 — Visualising the Distribution Shift

<div style='background:#fff1f2;border-left:6px solid #e11d48;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;'>
<b style='color:#9f1239;font-size:15px;'>Side-by-side feature histograms — train vs test</b><br><br>
Each of the 10 features is plotted as overlapping histograms (train in blue, test in orange). The 3 shifted features will show visibly different distributions — different means, spreads, or shapes. In practice you <em>would not know which features are shifted</em> in advance; this plot is shown here for ground-truth confirmation before running the adversarial classifier.
</div>

```python
fig, axes = plt.subplots(2, 5, figsize=(18, 6))
axes = axes.flatten()

for i, col in enumerate(X_train.columns):
    ax = axes[i]
    ax.hist(X_train[col], bins=40, alpha=0.6, label="Train", color="#4C72B0", density=True)
    ax.hist(X_test[col],  bins=40, alpha=0.6, label="Test",  color="#DD8452", density=True)
    ax.set_title(col, fontsize=10)
    ax.set_xlabel("")
    if col in shifted_features:
        ax.set_facecolor("#fff3f3")
        ax.set_title(f"⚠ {col}", fontsize=10, color="darkred")

axes[0].legend(fontsize=9)
plt.suptitle("Feature Distributions: Train vs Test  (red background = shifted)", fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig("distributions.png", dpi=120, bbox_inches="tight")
plt.show()
print("Notice features 0–2 have very different distributions in train vs test.")
```

## Section 4 — Running Adversarial Validation

<div style='background:#fdf4ff;border-left:6px solid #9333ea;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;margin-bottom:16px;'>
<b style='color:#6b21a8;font-size:15px;'>The algorithm — 4 steps</b>
<ol style='margin-top:10px;padding-left:18px;line-height:2.2;'>
  <li>Label all train rows as <code style='background:#ede9fe;padding:2px 5px;border-radius:4px;'>0</code> and all test rows as <code style='background:#ede9fe;padding:2px 5px;border-radius:4px;'>1</code></li>
  <li>Concatenate them into a single combined dataset</li>
  <li>Train a binary classifier to predict the label (is this row from train or test?)</li>
  <li>Evaluate with <b>stratified cross-validated AUC</b> — the higher it is, the more separable the sets are</li>
</ol>
</div>

<div style='background:#f8fafc;border-left:6px solid #475569;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;'>
<b style='color:#1e293b;font-size:15px;'>Interpreting the AUC</b>
<table style='border-collapse:collapse;width:100%;font-size:0.92em;background:white;border-radius:8px;overflow:hidden;margin-top:10px;'>
<thead><tr style='background:#475569;color:white;'>
<th style='padding:10px 14px;text-align:left;'>AUC Range</th>
<th style='padding:10px 14px;text-align:left;'>Interpretation</th>
<th style='padding:10px 14px;text-align:left;'>Action</th>
</tr></thead>
<tbody>
<tr style='background:#fff;border-bottom:1px solid #e2e8f0;'><td style='padding:9px 14px;font-weight:600;color:#16a34a;'>≈ 0.50</td><td style='padding:9px 14px;'>Train ≈ Test — no detectable shift</td><td style='padding:9px 14px;'>✅ No action needed</td></tr>
<tr style='background:#f8fafc;border-bottom:1px solid #e2e8f0;'><td style='padding:9px 14px;font-weight:600;color:#d97706;'>0.60 – 0.75</td><td style='padding:9px 14px;'>Mild shift — worth investigating</td><td style='padding:9px 14px;'>🔍 Inspect importances</td></tr>
<tr style='background:#fff;border-bottom:1px solid #e2e8f0;'><td style='padding:9px 14px;font-weight:600;color:#ea580c;'>0.75 – 0.90</td><td style='padding:9px 14px;'>Significant shift — likely to hurt performance</td><td style='padding:9px 14px;'>⚠️ Drop or transform features</td></tr>
<tr style='background:#f8fafc;'><td style='padding:9px 14px;font-weight:600;color:#e11d48;'>&gt; 0.90</td><td style='padding:9px 14px;'>Severe shift — model will generalise poorly</td><td style='padding:9px 14px;'>🚨 Fix upstream data issues</td></tr>
</tbody>
</table>
</div>

```python
def adversarial_validation(X_train, X_test, classifier=None, n_splits=5, verbose=True):
    """
    Run adversarial validation between train and test sets.
    
    Parameters
    ----------
    X_train    : pd.DataFrame — training features
    X_test     : pd.DataFrame — test features
    classifier : sklearn classifier (default: GradientBoostingClassifier)
    n_splits   : int — number of CV folds
    verbose    : bool — print results
    
    Returns
    -------
    auc_scores : list of per-fold AUC scores
    clf        : fitted classifier on the full adversarial dataset
    X_adv      : the combined adversarial DataFrame
    y_adv      : the combined adversarial labels
    """
    if classifier is None:
        classifier = GradientBoostingClassifier(
            n_estimators=100, max_depth=4, learning_rate=0.1, random_state=42
        )
    
    # Step 1 & 2: Label and combine
    X_adv = pd.concat([X_train, X_test], axis=0, ignore_index=True)
    y_adv = np.array([0] * len(X_train) + [1] * len(X_test))
    
    # Step 3 & 4: Cross-validated AUC
    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    auc_scores = cross_val_score(classifier, X_adv, y_adv, cv=cv, scoring="roc_auc")
    
    # Fit on full data for feature importance analysis
    clf = GradientBoostingClassifier(
        n_estimators=100, max_depth=4, learning_rate=0.1, random_state=42
    )
    clf.fit(X_adv, y_adv)
    
    if verbose:
        print("=" * 45)
        print("       ADVERSARIAL VALIDATION RESULTS")
        print("=" * 45)
        for fold, auc in enumerate(auc_scores, 1):
            bar = "█" * int(auc * 30)
            print(f"  Fold {fold}: AUC = {auc:.4f}  {bar}")
        print("-" * 45)
        print(f"  Mean AUC : {auc_scores.mean():.4f}")
        print(f"  Std  AUC : {auc_scores.std():.4f}")
        print("=" * 45)
        
        mean = auc_scores.mean()
        if mean < 0.55:
            verdict = "✅  No significant shift detected."
        elif mean < 0.70:
            verdict = "⚠️   Mild shift. Investigate top features."
        elif mean < 0.85:
            verdict = "🔶  Significant shift. Action recommended."
        else:
            verdict = "🚨  Severe shift! Model likely to fail."
        print(f"\nVerdict: {verdict}")
    
    return auc_scores, clf, X_adv, y_adv

auc_scores, clf, X_adv, y_adv = adversarial_validation(X_train, X_test)
```

## Section 5 — Diagnosing Which Features Cause the Shift

<div style='background:#fff1f2;border-left:6px solid #e11d48;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;'>
<b style='color:#9f1239;font-size:15px;'>Feature importances as a diagnostic tool</b><br><br>
Once a high AUC confirms shift exists, the next question is: <em>which features are responsible?</em> The adversarial classifier answers this directly — features that are most useful for <b>telling train from test apart</b> are the most shifted ones. Plotting the classifier's feature importances in descending order immediately pinpoints the culprits. Features known to be shifted (ground truth) are highlighted for comparison.
</div>

```python
def plot_adversarial_feature_importance(clf, feature_names, shifted_features=None, top_n=10):
    """Plot feature importances from the adversarial classifier."""
    importances = clf.feature_importances_
    indices = np.argsort(importances)[::-1][:top_n]
    
    colors = []
    for i in indices:
        if shifted_features and feature_names[i] in shifted_features:
            colors.append("#e74c3c")   # red = truly shifted
        else:
            colors.append("#3498db")   # blue = not shifted
    
    fig, ax = plt.subplots(figsize=(10, 5))
    bars = ax.barh(
        range(len(indices)),
        importances[indices][::-1],
        color=colors[::-1],
        edgecolor="white",
        linewidth=0.5
    )
    ax.set_yticks(range(len(indices)))
    ax.set_yticklabels([feature_names[i] for i in indices[::-1]], fontsize=11)
    ax.set_xlabel("Feature Importance (adversarial classifier)", fontsize=11)
    ax.set_title("Adversarial Feature Importances
(🔴 red = truly shifted features)", fontsize=13)
    ax.axvline(x=1/len(feature_names), color="grey", linestyle="--", alpha=0.7, label="Uniform baseline")
    ax.legend()
    
    # Add importance values
    for bar, idx in zip(bars[::-1], indices[::-1]):
        ax.text(bar.get_width() + 0.002, bar.get_y() + bar.get_height()/2,
                f"{importances[idx]:.3f}", va="center", fontsize=9)
    
    plt.tight_layout()
    plt.savefig("feature_importance.png", dpi=120, bbox_inches="tight")
    plt.show()

plot_adversarial_feature_importance(
    clf, 
    feature_names=list(X_train.columns),
    shifted_features=shifted_features
)
print("The adversarial classifier correctly ranks the shifted features highest!")
```

## Section 6 — ROC Curve Analysis

<div style='background:#eef2ff;border-left:6px solid #4f46e5;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;'>
<b style='color:#3730a3;font-size:15px;'>A richer view of classifier separability</b><br><br>
The single cross-validated AUC summarises performance in one number, but the <b>ROC curve</b> shows the full trade-off between true positive rate and false positive rate at every threshold. A curve hugging the top-left corner indicates the classifier easily separates train from test — strong evidence of shift. A curve close to the diagonal (random) means the sets are indistinguishable. Plotting both the shifted and no-shift baselines on the same axes makes the contrast stark.
</div>

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc

X_tr, X_val, y_tr, y_val = train_test_split(X_adv, y_adv, test_size=0.3, stratify=y_adv, random_state=42)

clf_plot = GradientBoostingClassifier(n_estimators=100, max_depth=4, random_state=42)
clf_plot.fit(X_tr, y_tr)
probs = clf_plot.predict_proba(X_val)[:, 1]

fpr, tpr, _ = roc_curve(y_val, probs)
roc_auc = auc(fpr, tpr)

fig, ax = plt.subplots(figsize=(7, 6))
ax.plot(fpr, tpr, color="#e74c3c", lw=2.5, label=f"Adversarial Classifier (AUC = {roc_auc:.3f})")
ax.plot([0, 1], [0, 1], color="gray", linestyle="--", lw=1.5, label="Random (AUC = 0.500)")
ax.fill_between(fpr, tpr, alpha=0.08, color="#e74c3c")
ax.set_xlabel("False Positive Rate", fontsize=12)
ax.set_ylabel("True Positive Rate", fontsize=12)
ax.set_title("Adversarial Validation ROC Curve", fontsize=14)
ax.legend(loc="lower right", fontsize=11)
ax.set_xlim([0, 1])
ax.set_ylim([0, 1.02])
ax.grid(alpha=0.3)

annotation = "Severe shift" if roc_auc > 0.85 else ("Significant shift" if roc_auc > 0.70 else "Mild/No shift")
ax.annotate(annotation, xy=(0.6, 0.2), fontsize=13, color="#e74c3c",
            fontweight="bold", ha="center")

plt.tight_layout()
plt.savefig("roc_curve.png", dpi=120, bbox_inches="tight")
plt.show()
```

## Section 7 — Mitigation Strategies

<div style='background:#f0fdf4;border-left:6px solid #16a34a;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;margin-bottom:16px;'>
<b style='color:#14532d;font-size:15px;'>Strategy A — Drop Shifted Features</b><br><br>
The simplest remedy: remove features where adversarial importance exceeds a threshold (default 10%). After dropping, adversarial validation is re-run to confirm the AUC has fallen. This approach trades predictive signal for distributional alignment — acceptable when the shifted features carry little label-predictive power, but costly when they don't.
</div>

```python
def drop_shifted_features(X_train, X_test, clf, feature_names, threshold=0.10):
    """
    Drop features where adversarial importance exceeds `threshold`.
    These are the features most responsible for train/test shift.
    """
    importances = dict(zip(feature_names, clf.feature_importances_))
    to_drop = [f for f, imp in importances.items() if imp > threshold]
    
    print(f"Features dropped (importance > {threshold}): {to_drop}")
    
    X_train_clean = X_train.drop(columns=to_drop)
    X_test_clean  = X_test.drop(columns=to_drop)
    
    return X_train_clean, X_test_clean, to_drop

X_train_clean, X_test_clean, dropped = drop_shifted_features(
    X_train, X_test, clf, list(X_train.columns), threshold=0.10
)

# Re-run adversarial validation to confirm improvement
print("\nAdversarial validation AFTER dropping shifted features:")
auc_clean, _, _, _ = adversarial_validation(X_train_clean, X_test_clean)
print(f"\nAUC improved from {auc_scores.mean():.4f} → {auc_clean.mean():.4f}")
```

<div style='background:#fffbeb;border-left:6px solid #d97706;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;'>
<b style='color:#92400e;font-size:15px;'>Strategy B — Sample Weighting (Importance Reweighting)</b><br><br>
Instead of discarding features, we <b>reweight training samples</b> so the training distribution better matches test. Each train sample receives a weight proportional to the <b>density ratio</b> <code style='background:#fef3c7;padding:2px 5px;border-radius:4px;'>P(test) / P(train)</code> — samples that "look like test" get higher weight, making the model pay more attention to them during training.<br><br>
The density ratio is estimated using the adversarial classifier's predicted probabilities: a train sample with high <code style='background:#fef3c7;padding:2px 5px;border-radius:4px;'>P(test | x)</code> looks test-like and deserves more weight. Extreme weights are clipped to prevent training instability.
</div>

```python
def compute_importance_weights(X_train, X_test, clip_percentile=99):
    """
    Compute density ratio weights: w(x) ∝ P(test) / P(train).
    
    Train the adversarial classifier, then use its predicted probability
    of being 'test' as a proxy for the density ratio.
    
    Parameters
    ----------
    X_train          : train features
    X_test           : test features
    clip_percentile  : clip extreme weights to avoid instability
    
    Returns
    -------
    weights : np.array of shape (n_train,)
    """
    X_adv = pd.concat([X_train, X_test], ignore_index=True)
    y_adv = np.array([0] * len(X_train) + [1] * len(X_test))
    
    clf = GradientBoostingClassifier(n_estimators=100, max_depth=4, random_state=42)
    clf.fit(X_adv, y_adv)
    
    # P(test | x) for train samples
    p_test = clf.predict_proba(X_train)[:, 1]   # prob of being test
    p_train = 1 - p_test                          # prob of being train
    
    # Density ratio: P(test)/P(train), clipped for stability
    weights = p_test / (p_train + 1e-6)
    clip_val = np.percentile(weights, clip_percentile)
    weights = np.clip(weights, 0, clip_val)
    weights = weights / weights.mean()             # normalise
    
    return weights

weights = compute_importance_weights(X_train, X_test)

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].hist(weights, bins=50, color="#3498db", edgecolor="white")
axes[0].set_title("Distribution of Importance Weights", fontsize=12)
axes[0].set_xlabel("Weight")
axes[0].set_ylabel("Count")
axes[0].axvline(1.0, color="red", linestyle="--", label="Weight = 1 (uniform)")
axes[0].legend()

# Show effect: weighted vs unweighted distribution of a shifted feature
feat = "feature_0"
axes[1].hist(X_train[feat], bins=40, alpha=0.5, label="Train (unweighted)", density=True, color="#4C72B0")
axes[1].hist(X_test[feat],  bins=40, alpha=0.5, label="Test",               density=True, color="#DD8452")
axes[1].hist(X_train[feat], bins=40, alpha=0.5, label="Train (reweighted)", density=True,
             color="#2ecc71", weights=weights)
axes[1].set_title(f"Effect of Reweighting on '{feat}'", fontsize=12)
axes[1].set_xlabel("Feature value")
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.savefig("weights.png", dpi=120, bbox_inches="tight")
plt.show()
print(f"Weight stats — mean: {weights.mean():.3f}, std: {weights.std():.3f}, max: {weights.max():.2f}")
```

## Section 8 — Comparing Adversarial Classifiers

<div style='background:#fdf4ff;border-left:6px solid #9333ea;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;'>
<b style='color:#6b21a8;font-size:15px;'>Does classifier choice matter?</b><br><br>
Different model families can give different signal quality on the same data. We benchmark three classifiers side by side:
<ul style='margin-top:8px;padding-left:18px;line-height:2;'>
  <li><b>Gradient Boosting</b> — typically the strongest signal; handles non-linear feature interactions well</li>
  <li><b>Random Forest</b> — robust and less prone to overfitting the adversarial labels</li>
  <li><b>Logistic Regression</b> — a linear baseline; a high AUC here means the shift is linearly detectable</li>
</ul>
Agreement across all three strengthens confidence that the detected shift is real rather than an artefact of one classifier's inductive bias.
</div>

```python
classifiers = {
    "GradientBoosting"  : GradientBoostingClassifier(n_estimators=100, max_depth=4, random_state=42),
    "RandomForest"      : RandomForestClassifier(n_estimators=100, max_depth=6, random_state=42),
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=42),
}

X_adv_all = pd.concat([X_train, X_test], ignore_index=True)
y_adv_all  = np.array([0] * len(X_train) + [1] * len(X_test))
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = {}
for name, clf_cmp in classifiers.items():
    scores = cross_val_score(clf_cmp, X_adv_all, y_adv_all, cv=cv, scoring="roc_auc")
    results[name] = scores
    print(f"{name:25s}  AUC = {scores.mean():.4f} ± {scores.std():.4f}")

fig, ax = plt.subplots(figsize=(9, 4))
names = list(results.keys())
means = [results[n].mean() for n in names]
stds  = [results[n].std()  for n in names]
colors_bar = ["#e74c3c", "#3498db", "#2ecc71"]

bars = ax.bar(names, means, yerr=stds, capsize=6, color=colors_bar, alpha=0.85, edgecolor="white")
ax.axhline(0.5, color="gray", linestyle="--", label="Random baseline (0.5)")
ax.set_ylabel("Cross-validated AUC", fontsize=11)
ax.set_title("Adversarial Validation AUC by Classifier", fontsize=13)
ax.set_ylim(0.4, 1.0)
ax.legend()
for bar, mean in zip(bars, means):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.015,
            f"{mean:.3f}", ha="center", fontsize=10, fontweight="bold")
plt.tight_layout()
plt.savefig("classifier_comparison.png", dpi=120, bbox_inches="tight")
plt.show()
```

## Section 9 — Sanity Check: No-Shift Baseline

<div style='background:#f0fdf4;border-left:6px solid #16a34a;padding:16px 20px;border-radius:8px;font-family:-apple-system,BlinkMacSystemFont,Segoe UI,sans-serif;color:#111;line-height:1.6;'>
<b style='color:#14532d;font-size:15px;'>Confirming the method returns ~0.5 when there is no shift</b><br><br>
A well-implemented adversarial validator should be calibrated — when train and test are drawn from <em>exactly</em> the same distribution, the classifier should perform no better than random (AUC ≈ 0.5). This cell generates a clean no-shift dataset and runs the full pipeline to confirm. Any result significantly above 0.5 on this check would indicate a bug in the implementation or data leakage.
</div>

```python
# Create a clean dataset with NO shift (same distribution)
X_no_shift_train = pd.DataFrame(np.random.randn(2000, 10), columns=[f"feature_{i}" for i in range(10)])
X_no_shift_test  = pd.DataFrame(np.random.randn(800,  10), columns=[f"feature_{i}" for i in range(10)])

print("Adversarial validation on IDENTICAL distributions (no shift):")
auc_no_shift, _, _, _ = adversarial_validation(X_no_shift_train, X_no_shift_test)

print("\nAdversarial validation on SHIFTED distributions (our original data):")
auc_shifted, _, _, _ = adversarial_validation(X_train, X_test, verbose=False)
print(f"  Mean AUC: {auc_shifted.mean():.4f}")

fig, ax = plt.subplots(figsize=(8, 4))
ax.boxplot([auc_no_shift, auc_shifted], labels=["No Shift", "With Shift"],
           patch_artist=True,
           boxprops=dict(facecolor="#dceefb"),
           medianprops=dict(color="#e74c3c", linewidth=2))
ax.axhline(0.5, color="gray", linestyle="--", label="Random baseline")
ax.axhline(0.7, color="orange", linestyle=":", label="Warning threshold (0.7)")
ax.set_ylabel("Cross-validated AUC")
ax.set_title("AUC Distribution: No Shift vs With Shift", fontsize=13)
ax.legend()
plt.tight_layout()
plt.savefig("sanity_check.png", dpi=120, bbox_inches="tight")
plt.show()
```
